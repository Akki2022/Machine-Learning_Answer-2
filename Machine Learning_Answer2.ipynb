```python
import pandas as pd

def get_category_counts(data):
    category_counts = {}
    for column in data.columns:
        categories = data[column].value_counts()
        category_counts[column] = {
            'Categories': list(categories.index),
            'Count': len(categories)
        }
    return pd.DataFrame(category_counts)

# Assuming your dataset is stored in a DataFrame called 'obesity_data'
category_counts_df = get_category_counts(data)

# Print the DataFrame

```

#FEATURE ENGINEERING

```python
data['NObeyesdad'].value_counts()
#This 7 types of categories are present in this column.
# So , we have to minimize the categories as this will be our output column and also it is an object type column , So we have to convert this into numerical type.
```

```python
category_mapping = {
    'Normal_Weight': 0,
    'Overweight_Level_I': 1,
    'Overweight_Level_II': 1,
    'Insufficient_Weight': 0,
    'Obesity_Type_I': 2,
    'Obesity_Type_II': 2,
    'Obesity_Type_III': 2
}

# Convert to numerical values and handle missing values
data['NObeyesdad'] = data['NObeyesdad'].map(category_mapping).fillna(-1)
```

```python
data.head()
```

```python
#Now, converting all the object type columns to numerical 
```

```python
# These are the columns Gender, family_history_with_overweight, FAVC, SMOKE	, SCC	which have two categories inside them
# So, for them we can use "One Hot Encoding for them"
```

```python
columns_to_encode = ['Gender','family_history_with_overweight','FAVC','SMOKE','SCC']  # Specify the columns to encode

data1 = pd.get_dummies(data, columns=columns_to_encode, drop_first=True)
```

```python
data1.head()
```

```python
# Now, the left columns are - 'CAEC', 'CALC',	'MTRANS	'
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data1[['CAEC', 'CALC', 'MTRANS']] = data1[['CAEC', 'CALC', 'MTRANS']].apply(lambda x: le.fit_transform(x))
```
